```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

base_url = "https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31"
driver.get(base_url)

wait = WebDriverWait(driver, 10)
```

```python
# -------------------------
# STEP 1: Click "50 more" until finished
# -------------------------
while True:
    try:
        show_more = wait.until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(.,'50 more')]"))
        )
        driver.execute_script("arguments[0].click();", show_more)
        print("Clicked 50 more...")
        time.sleep(3)
    except:
        print("No more '50 more' button found.")
        break
```

```python
# -------------------------
# STEP 2: Extract Movie Name + Storyline from Cards
# -------------------------

# Wait until movie cards load
wait.until(
    EC.presence_of_all_elements_located(
        (By.CSS_SELECTOR, "li.ipc-metadata-list-summary-item")
    )
)

movie_cards = driver.find_elements(
    By.CSS_SELECTOR,
    "li.ipc-metadata-list-summary-item"
)

print("Movie cards found:", len(movie_cards))

movie_names = []
storylines = []

for card in movie_cards:
    try:
        # Movie title
        title = card.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text

        # Short storyline from search page
        try:
            storyline = card.find_element(
                By.CSS_SELECTOR,
                "div.ipc-html-content-inner-div"
            ).text
        except:
            storyline = "Not Available"

        movie_names.append(title)
        storylines.append(storyline)

        print(f"Scraped: {title}")

    except Exception as e:
        print("Error:", e)
        continue
```

```python
# -------------------------
# STEP 3: Save Data
# -------------------------

df = pd.DataFrame({
    "Movie Name": movie_names,
    "Storyline": storylines
})

df.to_csv("imdb_2024_movies.csv", index=False)

print("Scraping Completed ✅")
print(f"Total Movies Scraped: {len(df)}")

driver.quit()
```
